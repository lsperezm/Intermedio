## Caracterización

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data = "data.xlsx"
data = pd.read_excel(data)
data
```

```python
V = data['kV']
V = V*1000
D = data['D2 (cm)']
D = D/100
Verr = data['error kV']
Verr = Verr*1000
Derr = data['Grosor D2 (cm)']
Derr = Derr/100
```

```python
X = 1/np.sqrt(V)
Xerr = ( 0.5/((V)**(3/2)) )*Verr

Y = D
Yerr = Derr


w = 1/(Yerr**2)

Xerr
```

```python
sum_w = w.sum()
sum_wx = (w*X).sum()
sum_wy = (w*Y).sum()
sum_wxx = (w*X**2).sum()
sum_wxy = (w*X*Y).sum()
```

```python
plt.scatter(X,Y)
plt.xlabel(r'$\frac{1}{\sqrt{V}}$ ($V^{\frac{-1}{2}}$)')
plt.ylabel(r'Diámetro $2^{do}$ círculo - D2 (m)')
plt.grid()
```

```python
m = (sum_w*sum_wxy - sum_wx*sum_wy) / (sum_w*sum_wxx - sum_wx**2)
b = (sum_wxx*sum_wy - sum_wx*sum_wxy) / (sum_w*sum_wxx - sum_wx**2)

m_std_err = np.sqrt(sum_w/(sum_w*sum_wxx-sum_wx**2))
b_std_err = np.sqrt(sum_wxx/(sum_w*sum_wxx-sum_wx**2))

model = m*X + b
```

```python
print(m)
print(b)

print(m_std_err)
print(b_std_err)
```

```python
fig1, (ax1) = plt.subplots(1, 1, figsize=(10, 10))  # 2 rows, 1 column, width = 10 inches, height = 10 inches

ax1.plot(X,model, color='red', linewidth=2, linestyle='--', label=f'D = ({m:.4f}± {m_std_err:.4f})' r'$V^{\frac{-1}{2}}$' f' + ({b:.3f}± {b_std_err:.3f})')
ax1.errorbar(X, Y, Yerr, Xerr,  fmt='o')

ax1.set_xlabel(r'$\frac{1}{\sqrt{V}}$ ($V^{\frac{-1}{2}}$)')
ax1.set_ylabel(r'Diámetro $2^{do}$ círculo - D2 (m)')
ax1.legend(loc='upper left')
ax1.set_title(r'$D_2$' ' en función de ' r'$V^{\frac{-1}{2}}$')
ax1.grid()

plt.savefig('plot_D2.png', format='png')
```

```python
res = (Y - model)/Yerr

fig2, (ax2) = plt.subplots(1, 1, figsize=(20, 10))

ax2.scatter(X, res)
ax2.axhline(0, color='black', linewidth=1, linestyle='--')

ax2.set_xlabel(r'$\frac{1}{\sqrt{V}}$ ($V^{\frac{-1}{2}}$)')
ax2.set_ylabel('Residuales')
ax2.set_title(r'$D_2$' ' contra ' r'$V^{\frac{-1}{2}}$' ' - Resiuales Normalizados')
ax2.grid()

plt.savefig('residuales_D2.png', format='png')
```

```python
def caracterizacion(I):
    
    return m*I + b
```

```python
lamb=632.816e-9
l=14.25e-2
delta=lambda N: N*lamb/2

sigmaL = lambda dl: (dl/(l**2))*0.00005
sigmaB = lambda I: np.sqrt( (I*m_std_err)**2 + (m*0.01)**2 + (b_std_err)**2 )
```

## Hierro

```python
I_Fe=np.array([0.14,1.18,2.55])
Conteo_Fe=np.array([0,1,0])
```

```python
plt.scatter(I_Fe,Conteo_Fe)
plt.xlabel('Corriente (A)')
plt.ylabel('Conteo de Mínimos')
plt.grid()
```

```python
B_Fe = caracterizacion(I_Fe)
deltaL_Fe = delta(Conteo_Fe)
Long_Fe = deltaL_Fe/l
sigmaL_Fe = sigmaL(deltaL_Fe)
sigmaB_Fe = sigmaB(I_Fe)
```

```python
plt.errorbar(B_Fe, Long_Fe, sigmaL_Fe, sigmaB_Fe,  fmt='o')
plt.xlabel('Campo Magnético [mT]')
plt.ylabel('ΔL/L')
plt.grid()
plt.title('Magnestostricción del Hierro')

plt.savefig('plot_hierro.png', format='png')
```
