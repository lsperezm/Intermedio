```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data = "data.xlsx"
data = pd.read_excel(data)
data
```

```python
V = data['kV']
V = V*1000
D = data['D1 (cm)']
D = D/100
Verr = data['error kV']
Verr = Verr*1000
Derr = data['Grosor D1 (cm)']
Derr = Derr/100
```

```python
X = 1/np.sqrt(V)
Xerr = ( 0.5/((V)**(3/2)) )*Verr

Y = D
Yerr = Derr


w = 1/(Yerr**2)

Xerr
```

```python
sum_w = w.sum()
sum_wx = (w*X).sum()
sum_wy = (w*Y).sum()
sum_wxx = (w*X**2).sum()
sum_wxy = (w*X*Y).sum()
```

```python
plt.scatter(X,Y)
plt.xlabel(r'$\frac{1}{\sqrt{V}}$ ($V^{\frac{-1}{2}}$)')
plt.ylabel(r'Diámetro $1^{er}$ círculo - D1 (m)')
plt.grid()
```

```python
m = (sum_w*sum_wxy - sum_wx*sum_wy) / (sum_w*sum_wxx - sum_wx**2)
b = (sum_wxx*sum_wy - sum_wx*sum_wxy) / (sum_w*sum_wxx - sum_wx**2)

m_std_err = np.sqrt(sum_w/(sum_w*sum_wxx-sum_wx**2))
b_std_err = np.sqrt(sum_wxx/(sum_w*sum_wxx-sum_wx**2))

model = m*X + b
```

```python
print(m)
print(b)

print(m_std_err)
print(b_std_err)
```

```python
fig1, (ax1) = plt.subplots(1, 1)

ax1.plot(X,model, color='red', linewidth=2, linestyle='--', label=r'$D_1$'f' = ({m:.1f}± {m_std_err:.1f})' r'$V^{\frac{-1}{2}}$' f' + ({b:.3f}± {b_std_err:.3f})')
ax1.errorbar(X, Y, Yerr, Xerr,  fmt='o')

ax1.set_xlabel(r'$\frac{1}{\sqrt{V}}$ ($V^{\frac{-1}{2}}$)')
ax1.set_ylabel(r'Diámetro $1^{er}$ círculo - D1 (m)')
ax1.legend(loc='upper left')
ax1.set_title(r'$D_1$' ' en función de ' r'$V^{\frac{-1}{2}}$')
ax1.grid()

plt.savefig('plot_D1.png', format='png')
```

```python
res = (Y - model)/Yerr

fig2, (ax2) = plt.subplots(1, 1)

ax2.scatter(X, res)
ax2.axhline(0, color='black', linewidth=1, linestyle='--')

ax2.set_xlabel(r'$\frac{1}{\sqrt{V}}$ ($V^{\frac{-1}{2}}$)')
ax2.set_ylabel('Residuales')
ax2.set_title(r'$D_1$' ' contra ' r'$V^{\frac{-1}{2}}$' ' - Resiuales Normalizados')
ax2.grid()

plt.savefig('residuales_D1.png', format='png')
```
