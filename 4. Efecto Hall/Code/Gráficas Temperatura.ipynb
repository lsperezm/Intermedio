## Placa Geranio N

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data_n = "Datos_Temp_n.xlsx"
data_n = pd.read_excel(data_n)
data_n
```

### Corriente=0A

```python
temp_0=np.array(data_n['T [°C]'])
VH_0=np.array(data_n['V_H [mV]'])*(-1)
VL_0=np.array(data_n['V_L [V]'])
VHerr_0=np.array(data_n['VH err'])
VLerr_0=np.array(data_n['VL err'])
terr=np.ones(len(temp_0))
```

```python
plt.errorbar(temp_0,VH_0,VHerr_0,terr,fmt='.')
plt.xlabel('Temperatura [°C]')
plt.ylabel('Voltaje Hall [V]')
plt.title('Placa N-Ge con Campo Magnético Nulo')
```

```python
plt.errorbar(temp_0,VL_0,VLerr_0,terr,fmt='.')

plt.xlabel('Temperatura [°C]')
plt.ylabel('Voltaje Longitudinal [V]')
plt.title('Placa N-Ge con Campo Magnético Nulo')
```

### Corriente= 1.4A

```python
temp_1=np.array(data_n['T [°C].1'])
VH_1=np.array(data_n['V_H [mV].1'])*(-1)
VL_1=np.array(data_n['V_L [V].1'])
VHerr_1=np.array(data_n['VH err.1'])
VLerr_1=np.array(data_n['VL err.1'])
```
