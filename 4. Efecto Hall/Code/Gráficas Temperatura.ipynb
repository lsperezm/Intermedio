## Placa Geranio N

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data_n = "Datos_Temp_n.xlsx"
data_n = pd.read_excel(data_n)
data_n
```

### Corriente=0A

```python
temp_0=np.array(data_n['T [°C]'])+273
VH_0=np.array(data_n['V_H [mV]'])*(-1)/1000
VL_0=np.array(data_n['V_L [V]'])
VHerr_0=np.array(data_n['VH err'])/1000
VLerr_0=np.array(data_n['VL err'])
terr=np.ones(len(temp_0))
```

```python
plt.errorbar(temp_0,VH_0,VHerr_0,terr,fmt='.')
plt.xlabel('Temperatura [°C]')
plt.ylabel('Voltaje Hall [V]')
plt.title('Placa N-Ge con Campo Magnético Nulo')
```

### Corriente= 1.4A

```python
temp_1=np.array(data_n['T [°C].1'])+273
VH_1=np.array(data_n['V_H [mV].1'])*(-1)/1000
VL_1=np.array(data_n['V_L [V].1'])
VHerr_1=np.array(data_n['VH err.1'])/1000
VLerr_1=np.array(data_n['VL err.1'])
```

```python
plt.errorbar(temp_1,VH_1,VHerr_1,terr,fmt='.')

plt.xlabel('Temperatura [°C]')
plt.ylabel('Voltaje Hall [V]')
plt.title('Placa N-Ge Campo Magnético de 318.78$\pm$0.08mT')
```

```python
I=0.03
w=1e-3
B=0.31878

Rh=lambda v: v*w/(B*I)
```

```python
Rh_N=Rh(VL_1)
```

```python
plt.scatter(temp_1,Rh_N)
plt.xlabel('Temperatura [°C]')
plt.ylabel('Coeficiente de Hall')
plt.title('Placa N-Ge Campo Magnético de 318.78$\pm$0.08mT')
```

```python
temp_1
```

```python
new_t=temp_1[40:]
new_rh_n= Rh_N[40:]
rh_err=lambda VH_err: VH_err*w/(B*I)
rh_nerr=rh_err(VHerr_1[40:])
```

```python
plt.scatter(new_t,new_rh_n)
```

```python
len(rh_nerr)
```

```python
log=lambda r,t:np.log(r*t**(3/2))
t_inv=1/new_t

log_N=log(new_rh_n,new_t)

t_err=lambda t:1/(t**2)
log_err=lambda rh_err,t,rh,t_err:np.sqrt(rh_err**2/rh**2 + (3*t_err)**2/(2*t)**2)

t_errN=t_err(new_t)
logN_err=log_err(rh_nerr,t_inv,new_rh_n,t_errN)
we = 1/(logN_err**2)
logN_err
```

```python
sum_w = we.sum()
sum_wx = (we*t_inv).sum()
sum_wy = (we*log_N).sum()
sum_wxx = (we*t_inv**2).sum()
sum_wxy = (we*t_inv*log_N).sum()
```

```python
mN = (sum_w*sum_wxy - sum_wx*sum_wy) / (sum_w*sum_wxx - sum_wx**2)
bN = (sum_wxx*sum_wy - sum_wx*sum_wxy) / (sum_w*sum_wxx - sum_wx**2)

m_std_Nerr = np.sqrt(sum_w/(sum_w*sum_wxx-sum_wx**2))
b_std_Nerr = np.sqrt(sum_wxx/(sum_w*sum_wxx-sum_wx**2))

model = mN*t_inv + bN
```

```python
fig1, (ax1) = plt.subplots(1, 1,figsize=(8,5))

ax1.plot(t_inv,model, color='red', linewidth=2, linestyle='--', label=r'$ln(R_H T^{3/2})$'f' = ({mN:.2f}± {m_std_Nerr:.2f})' r'I' f' + ({bN:.2f}± {b_std_Nerr:.2f})')
ax1.errorbar(t_inv, log_N, logN_err, t_errN,  fmt='o')

ax1.set_xlabel(r'$T^{-1}$')
ax1.set_ylabel(r'$ln(R_HT^{3/2})$')
ax1.legend(loc='upper left')
ax1.set_title('Placa Geranio N')
ax1.grid()

plt.savefig('tempGeN.png', format='png')
```

```python
K_B=8.617e-5
E_N=2*K_B*mN
E_err=2*K_B*m_std_Nerr
E_N,E_err
```

```python
res = (log_N - model)/logN_err

fig2, (ax2) = plt.subplots(1, 1,figsize=(8,3))

ax2.scatter(t_inv, res)
ax2.axhline(0, color='green', linewidth=1, linestyle='--')

ax2.set_xlabel(r'$T^{-1}$')
ax2.set_ylabel(r'$ln(R_HT^{3/2})$')
ax2.set_title('Temperatura Placa N - Residuales Normalizados')
ax2.grid()

plt.savefig('residuales_energiaN.png', format='png')
```

```python
rh_ierr=rh_err(VHierr_1)
```

```python
ti_inv=1/tempi_1

log_I=log(RH_I,tempi_1)

t_erri=t_err(tempi_1)
logi_err=log_err(rh_ierr,ti_inv,RH_I,t_erri)
wi = 1/(logi_err**2)
log_I
```

```python
sum_wi = wi.sum()
sum_wxi = (wi*ti_inv).sum()
sum_wyi = (wi*log_I).sum()
sum_wxxi = (wi*ti_inv**2).sum()
sum_wxyi = (wi*ti_inv*log_I).sum()
```

```python
mi = (sum_wi*sum_wxyi - sum_wxi*sum_wyi) / (sum_wi*sum_wxxi - sum_wxi**2)
bi = (sum_wxxi*sum_wyi - sum_wxi*sum_wxyi) / (sum_wi*sum_wxxi - sum_wxi**2)

m_std_ierr = np.sqrt(sum_wi/(sum_wi*sum_wxxi-sum_wxi**2))
b_std_ierr = np.sqrt(sum_wxxi/(sum_wi*sum_wxxi-sum_wxi**2))

modeli = mi*ti_inv + bi
modeli
```

```python
fig1, (ax1) = plt.subplots(1, 1,figsize=(8,5))

ax1.plot(ti_inv,modeli, color='red', linewidth=2, linestyle='--', label=r'$ln(R_H T^{3/2})$'f' = ({mi:.2f}± {m_std_ierr:.2f})' r'I' f' + ({bi:.2f}± {b_std_ierr:.2f})')
ax1.errorbar(ti_inv, log_I, logi_err, t_erri,  fmt='o')

ax1.set_xlabel(r'$T^{-1}$')
ax1.set_ylabel(r'$ln(R_HT^{3/2})$')
ax1.legend(loc='upper left')
ax1.set_title('Placa Geranio Intrínseco')
ax1.grid()

plt.savefig('tempGeP.png', format='png')
```
