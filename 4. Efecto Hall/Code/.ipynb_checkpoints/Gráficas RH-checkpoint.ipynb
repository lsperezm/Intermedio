```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
RH_Bcte_p = [0.00474, 0.00470, 0.00469, 0.00467, 0.00474]
RH_Bcte_n = [-0.00667, -0.00647, -0.00652, -0.00667, -0.00660]

B_array = [170.22, 181.65, 193.08, 204.5, 215.9]

RH_Icte_p = [0.0048, 0.0058, 0.0039, 0.0043]
RH_Icte_n = [-0.0063, -0.0065, -0.0051, -0.0054]

I_array = [-20, -10, 10, 20]
```

```python
RH_Bcte_p_prom = 0.004706
RH_Bcte_n_prom = -0.006586

RH_Icte_p_prom = 0.0047
RH_Icte_n_prom = -0.0058
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 6))

ax1.set_xlabel("Campo Magnético (mT)")
ax1.set_ylabel("RH")
ax1.set_title("Coeficiente de Hall en función del Campo Magnético")

ax1.axhline(RH_Bcte_p_prom, color='orange', linestyle='--', label='RH promedio = ' + str(RH_Bcte_p_prom) + ' (Germanio p)')
ax1.axhline(RH_Bcte_n_prom, color='magenta', linestyle='--', label='RH promedio = ' + str(RH_Bcte_n_prom) + ' (Germanio n)')
ax1.scatter(B_array, RH_Bcte_p)
ax1.scatter(B_array, RH_Bcte_n)
ax1.set_ylim(-0.007, 0.007)
ax1.legend()


ax2.set_xlabel("Corriente (mA)")
ax2.set_ylabel("RH")
ax2.set_title("Coeficiente de Hall en función de la Corriente")

ax2.axhline(RH_Icte_p_prom, color='orange', linestyle='--', label='RH promedio = ' + str(RH_Icte_p_prom) + ' (Germanio p)')
ax2.axhline(RH_Icte_n_prom, color='magenta', linestyle='--', label='RH promedio = ' + str(RH_Icte_n_prom) + ' (Germanio n)')
ax2.scatter(I_array, RH_Icte_p)
ax2.scatter(I_array, RH_Icte_n)
ax2.set_ylim(-0.007, 0.007)
ax2.legend()


plt.tight_layout()

plt.savefig("Graficas RH.png", format = "png")
```
