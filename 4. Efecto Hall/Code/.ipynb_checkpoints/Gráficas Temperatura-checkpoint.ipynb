## Placa Geranio N

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data_n = "Datos_Temp_n.xlsx"
data_n = pd.read_excel(data_n)
data_n
```

### Corriente=0A

```python
temp_0=np.array(data_n['T [°C]'])
VH_0=np.array(data_n['V_H '])
VL_0=np.array(data_n['V_L'])
```

```python
plt.scatter(temp_0,VH_0)
plt.xlabel('Temperatura [°C]')
plt.ylabel('Voltaje Hall [V]')
```

```python
plt.scatter(temp_0,VL_0)

plt.xlabel('Temperatura [°C]')
plt.ylabel('Voltaje Longitudinal [V]')
```

### Corriente= 1.4A

```python
temp_1=np.array(data_n['T [°C].1'])
VH_1=np.array(data_n['V_H .1'])
VL_1=np.array(data_n['V_L.1'])
```
