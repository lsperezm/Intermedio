```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data = "Datos_VL_p.xlsx"
data = pd.read_excel(data)
data
```

```python
X = data['B_rounded (mT)']
Xerr = data['B_err_rounded (mT)']
```

Ip = -20mA

```python
Y = data['VL_-20mA (V)']
Yerr = data['VL_err (V)']
```

```python
R0 = 0.732/(20/1000)

Rm = Y/(20/1000)
Rm_err = Yerr/(20/1000)

plot1 = ( (Rm-R0)/R0 )
err1 = Rm_err/R0

plt.errorbar(X, plot1, err1, Xerr, fmt='o')
plt.xlabel('Campo Magn√©tico (mT)')
plt.ylabel(r'$\frac{R_m-R_0}{R_0}$')
plt.title('Magnetorresistencia con ' + r'$I_p=-20mA$' + ' (Germanio p)')

plt.savefig('Magnetorresistencia -20mA_p.png', format='png')
```

Escogemos sets de datos adecuados para los ajustes.

```python
X_neg = -data['B (mT)'][:7]
X_neg
```

```python
X_pos = data['B (mT)'][-7:]
X_pos
```

```python
datos1 = plot1[:7]
datos2 = plot2[:7]
datos3 = plot3[:7]
datos4 = plot4[-7:]
```

```python
def fit(B, a, b, e):
    return a*(B**b) + e

X_test = np.linspace(0, 320, 32001)
```

```python
popt_1, pcov_1 = curve_fit( fit, X_neg, datos1, sigma=err1[:7], absolute_sigma=True, maxfev = 10000 )
```

```python
print("a: " + str(popt_1[0]) + " +/- " + str(np.sqrt(pcov_1[0, 0])))
print("b: " + str(popt_1[1]) + " +/- " + str(np.sqrt(pcov_1[1, 1])))
print("e: " + str(popt_1[2]) + " +/- " + str(np.sqrt(pcov_1[2, 2])))
```

```python
Y_fit1 = fit(X_test, *popt_1)

plt.xlabel("Campo Magn√©tico (mT)")
plt.ylabel(r"$R_m-R_0/R_0$")
plt.title("Magnetorresistencia con " + r'$I_p = -20mA$' + " (Germanio p)")

plt.errorbar(X_neg, datos1, err1[:7], Xerr[:7], fmt='o')
plt.plot(X_test, Y_fit1, label=r'$( (4\pm 7)\times 10^{-7} )B^{1.9\pm 0.3} + (-0.006\pm 0.001)$')
plt.legend(loc='upper left')

plt.savefig('Magnetorresistencia -20mA_p.png', format='png')
```
