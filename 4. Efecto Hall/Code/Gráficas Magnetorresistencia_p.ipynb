```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

data = "Datos_VL_p.xlsx"
data = pd.read_excel(data)
data
```

```python
X = data['B_rounded (mT)']
Xerr = data['B_err_rounded (mT)']
```

```python
w = 0.001
d = 0.01
l = 0.02

A = w*d
```

Ip = -20mA

```python
Y = data['VL_-20mA (V)']
Yerr = data['VL_err (V)']
```

```python
R0 = 0.732/(20/1000)

Rm = Y/(20/1000)
sigma1 = l/(Rm*A)
Rm_err = Yerr/(20/1000)

plot1 = ( (Rm-R0)/R0 )
err1 = Rm_err/R0

plt.errorbar(X, plot1, err1, Xerr, fmt='o')
plt.xlabel('Campo Magnético (mT)')
plt.ylabel(r'$\frac{R_m-R_0}{R_0}$')
plt.title('Magnetorresistencia con ' + r'$I_p=-20mA$' + ' (Germanio p)')

plt.savefig('Magnetorresistencia -20mA_p.png', format='png')
```

Escogemos sets de datos adecuados para los ajustes.

```python
X_neg = -data['B (mT)'][:7]
X_neg
```

```python
X_pos = data['B (mT)'][-7:]
X_pos
```

```python
fig1, ax1 = plt.subplots(2, 2, figsize = (12, 12))
```

```python
datos1 = plot1[:7]
datos2 = plot2[:7]
datos3 = plot3[:7]
datos4 = plot4[-7:]
```

```python
def fit(B, a, b, e):
    return a*(B**b) + e

X_test = np.linspace(0, 320, 32001)
```

```python
popt_1, pcov_1 = curve_fit( fit, X_neg, datos1, sigma=err1[:7], absolute_sigma=True, maxfev = 10000 )
```

```python
print("a: " + str(popt_1[0]) + " +/- " + str(np.sqrt(pcov_1[0, 0])))
print("b: " + str(popt_1[1]) + " +/- " + str(np.sqrt(pcov_1[1, 1])))
print("e: " + str(popt_1[2]) + " +/- " + str(np.sqrt(pcov_1[2, 2])))
```

```python
Y_fit1 = fit(X_test, *popt_1)

ax1[0][0].set_xlabel("Campo Magnético (mT)")
ax1[0][0].set_ylabel(r"$R_m-R_0/R_0$")
ax1[0][0].set_title("Magnetorresistencia con " + r'$I_p = -20mA$' + " (Germanio p)")

ax1[0][0].errorbar(X_neg, datos1, err1[:7], Xerr[:7], fmt='o')
ax1[0][0].plot(X_test, Y_fit1, label=r'$( (4\pm 7)\times 10^{-7} )B^{1.9\pm 0.3} + (-0.006\pm 0.001)$')
ax1[0][0].legend(loc='upper left')
```

```python
res = ( fit(X_neg, *popt_1) - datos1)/err1[:7]

plt.scatter(X_neg, res)
plt.axhline(0, color='green', linewidth=1, linestyle='--')

plt.xlabel('Campo Magnético (mT)')
plt.ylabel('Residuales normalizados')
plt.title("Residuales para Magnetorresistencia con " + r'$I_p = -20mA$' + " (Germanio p)")
plt.grid()

plt.savefig('Mag_p_-20mA_res.png', format='png')
```
