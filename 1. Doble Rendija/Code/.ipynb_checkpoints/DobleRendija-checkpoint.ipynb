```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import pandas as pd
import os.path as path
import os
```

# Láser
### Doble Rendija

```python
data_doble=pd.read_csv('Doble.csv')
data_doble
```

```python
x=np.array(data_doble.x)
V=np.array(data_doble.V)
xs=[]
Vs=[]
for i in range(len(x)):
    xi=float(x[i])
    xs.append(xi)
    Vi=float(V[i])
    Vs.append(Vi)
```

```python
plt.scatter(xs,Vs,marker='.')
plt.xlabel('x ($\mu m$)')
plt.ylabel('Voltaje (mV)')
```
